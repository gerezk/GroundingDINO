```python
from groundingdino.util.inference import load_model, load_image, predict, annotate
from pathlib import Path
import cv2
```

```python
model = load_model("../groundingdino/config/GroundingDINO_SwinT_OGC.py", "../weights/groundingdino_swint_ogc.pth", device="cpu")
BOX_TRESHOLD = 0.35
TEXT_TRESHOLD = 0.25
experiment_io_path = Path("../experiment_io/")
```

# Semantic Knowledge vs Visual Features

```python
image_path = "../experiment_io/images/apple.jpg"
image_name = Path(image_path).stem
image_extension = Path(image_path).suffix
text_prompts = ["apple", "fruit", "food", "round object", "red object"]

image_source, image = load_image(image_path)

for text_prompt in text_prompts:
    boxes, logits, phrases = predict(
        model=model,
        image=image,
        caption=text_prompt,
        box_threshold=BOX_TRESHOLD,
        text_threshold=TEXT_TRESHOLD,
        device="cpu",
    )

    print(phrases, logits)
    output_path = experiment_io_path / "predictions" / (image_name + " " + text_prompt + image_extension)
    annotated_frame = annotate(image_source=image_source, boxes=boxes, logits=logits, phrases=phrases)
    cv2.imwrite(output_path, annotated_frame)
```
